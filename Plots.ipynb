```python
import pandas as pd
df = pd.read_csv('Heart_disease_cleveland_new.csv') 
df.head()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df['chol'],kde=True)
plt.title("Cholesterol Distribution")
plt.show()
```

```python
sns.histplot(data=df, x='chol', hue='target', kde="True")
```

```python
sns.histplot(data=df, x='age', hue='target', kde=True)
```

```python
sns.boxplot(x=df['trestbps'])
plt.title("Resting Blood Pressure")
plt.show()
```

```python
sns.boxplot(data=df, x='chol')
plt.title("Cholesterol")
plt.show()
```

```python
sns.boxplot(x='target', y='chol', data=df)
plt.title("Target and Cholesterol")
plt.show() 
```

```python
sns.barplot(data=df, x='target', y='age')
plt.title("Target and Age")
plt.show() 
```

```python
sns.barplot(data=df, x="target", y="age", hue="target")
plt.title("Target and Age Distribution")
plt.xlabel("Target Categories")
plt.ylabel("Average Age")
plt.show()
```

```python
sns.countplot(data=df, x='target')
```

```python
sns.scatterplot(data=df, x='age', y='chol', hue="age" )
```

```python
sns.lineplot(data=df, x=df.index, y='thalach')
plt.title("Heart rate values change across dataset rows")
plt.show() 
```

```python
sns.lineplot(data=df.sort_values('age'), x='age', y='thalach')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr = df.corr()
plt.figure(figsize=(12, 8)) 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
